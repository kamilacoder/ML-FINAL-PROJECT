
<h2 style="color: violet;">Identify and Handle Outliers:</h2>


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.boxplot(data=df[['Likes', 'Platform', 'Hour', 'Retweets']])
plt.title("Boxplots for Score Distribution")
plt.show()
```

<h2 style="color: violet;">Calculate the Interquartile Range(IQR)  &  Determine the lower and upper bounds for IQR</h2>


```python
# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df[['Likes', 'Hour', 'Retweets']].quantile(0.25)
Q3 = df[['Likes', 'Hour', 'Retweets']].quantile(0.75)

# Compute IQR'Likes', 'Hour', 'Retweets'
IQR = Q3 - Q1
print("Interquartile Range (IQR):\n", IQR)
```

```python
# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:\n", lower_bound)
print("Upper Bound:\n", upper_bound)
```

```python
filtered_df = df[~((df[['Likes', 'Hour', 'Retweets']] < lower_bound) | 
                   (df[['Likes', 'Hour', 'Retweets']] > upper_bound)).any(axis=1)]

print("Original Dataset Size:", df.shape)
print("Filtered Dataset Size (after outlier removal):", filtered_df.shape)
```

<h2 style="color: violet;">Handle the outliers by removing them from the dataset.</h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Assume df is your DataFrame containing the score columns
score_columns = ['Likes', 'Hour', 'Retweets', 'Platform']

for col in score_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Math score histogram
sns.histplot(df["Likes"], kde=True, bins=20, edgecolor='black', color="teal", ax=axes[0])
axes[0].set_title('Removing outliers for Likes', fontsize=16, fontweight="bold")

# Math score histogram
sns.histplot(df["Platform"], kde=True, bins=20, edgecolor='black', color="teal", ax=axes[1])
axes[1].set_title('Removing outliers for Platform', fontsize=16, fontweight="bold")

plt.tight_layout()
plt.show()
```

<h2 style="color: violet;">Discuss whether the distributions are skewed to the left or right and suggest any appropriate transformations if needed to address skewness</h2>


```python
def anlys_skewness(data):
    skewness = data.skew()
    print("\n")
    if skewness > 0:
        print("Positively skewed (right skewed).")
        print("Apply a log or square root .")
    elif skewness < 0:
        print("Negatively skewed (left skewed).")
        print("Apply a reflection and then log or square root.")
    else:
        print("This is approximately symmetrical.")
        print("No transformation needed.")


anlys_skewness(df['Likes'])
anlys_skewness(df['Retweets'])
anlys_skewness(df['Hour'])
```

```python
m_log =df.copy()
m_log["LogM_alt"] = np.log1p(m_log["Likes"])
r_log =df.copy()
r_log["LogR_alt"] = np.log1p(r_log["Retweets"])
w_log =df.copy()
w_log["LogW_alt"] = np.log1p(w_log["Year"])
```
